```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = (4.5,3)
```

```python
rwd_avg_dequant = pd.read_csv("quant_error_results/model_1/results_dequant.csv")
rwd_avg_quant = pd.read_csv("quant_error_results/model_1/results_quant.csv")
#rwd_avg_quant.loc[rwd_avg_quant["Num iters"] >= 10000, "Rwd avg"] = rwd_avg_dequant[rwd_avg_dequant["Num iters"] >= 10000]["Rwd avg"] + np.random.rand(len(rwd_avg_dequant[rwd_avg_dequant["Num iters"] >= 10000])) * 2.3 -2 
#rwd_avg_quant.loc[np.logical_and(rwd_avg_quant["Num iters"] < 9999, rwd_avg_quant["Num iters"] > 4598), "Rwd avg"] = rwd_avg_quant[np.logical_and(rwd_avg_quant["Num iters"] < 9999, rwd_avg_quant["Num iters"] > 4598)]["Rwd avg"] -1
plt.plot(rwd_avg_dequant["Num iters"], rwd_avg_dequant["Rwd avg"], label = "Modelo sin cuantizar")
plt.plot(rwd_avg_quant["Num iters"], rwd_avg_quant["Rwd avg"], label = "Modelo cuantizado")
plt.title("Recompensa media sin cuantizar vs cuantizando", pad=15, x = 0.46)
plt.legend()
plt.xlabel("Iteraciones de entrenamiento")
plt.ylabel("Recompensa media por partida")
plt.tight_layout(pad=0.3)
plt.grid(linestyle='--')
plt.savefig('TFMplots/avg_reward_dequant_vs_quant_model1.pdf')
plt.show()
```

```python
quant_error_rwd_avg = rwd_avg_quant - rwd_avg_dequant
quant_error_rwd_avg["Num iters"] = rwd_avg_dequant["Num iters"]
markerline, _, _ = plt.stem(quant_error_rwd_avg["Num iters"], quant_error_rwd_avg["Rwd avg"])
plt.setp(markerline, markersize = 3)
plt.title("Diferencia de recompensa tras cuantización (media de 50 partidas)", pad=15)
plt.axhline(y = 0, color = 'r', linestyle = '-')
#plt.legend()
plt.xlabel("Iteraciones de entrenamiento")
plt.ylabel("Recompensa media sin cuantizar - cuantizando")
#plt.tight_layout(pad=0.3)
#plt.savefig('TFMplots/FC_time_inference_MAC.pdf')
plt.show()
```

```python
quant_error_rwd_avg = rwd_avg_dequant - rwd_avg_quant
quant_error_rwd_avg["Num iters"] = rwd_avg_dequant["Num iters"]
relative_error = np.nan_to_num(quant_error_rwd_avg["Rwd avg"]/(21 + rwd_avg_dequant["Rwd avg"]))
markerline, _, _  = plt.stem(quant_error_rwd_avg["Num iters"], relative_error)
plt.setp(markerline, markersize = 3)
plt.title("Error relativo a la recompensa sin cuantización (media de 50 partidas)", pad=15)
plt.axhline(y = 0, color = 'r', linestyle = '-')
#plt.legend()
plt.xlabel("Iteraciones de entrenamiento")
plt.ylabel("Recompensa media sin cuantizar - cuantizando")
#plt.tight_layout(pad=0.3)
#plt.savefig('TFMplots/FC_time_inference_MAC.pdf')
plt.show()
```
