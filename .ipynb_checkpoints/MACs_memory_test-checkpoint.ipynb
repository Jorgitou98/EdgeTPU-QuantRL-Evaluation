### TPU vs CPU

```python
cpu_8FC = pd.read_csv("FC_MACs/results/cpu-minN100-maxN2640-stepN40-threads8.csv")
cpu_4FC = pd.read_csv("FC_MACs/results/cpu-minN100-maxN2640-stepN40-threads4.csv")
cpu_2FC = pd.read_csv("FC_MACs/results/cpu-minN100-maxN2640-stepN40-threads2.csv")
cpu_1FC = pd.read_csv("FC_MACs/results/cpu-minN100-maxN2640-stepN40-threads1.csv")
tpu_FC = pd.read_csv("FC_MACs/results/minN100-maxN2640-stepN40-L4-I64.csv")
fig,ax = plt.subplots()
plt.title("Tiempo en Edge TPU y CPU. (Capas densas)", pad=15)
ax.set_yticks(np.arange(0, 40, 5))
plt.xticks(np.arange(0*10**7, 2.5*10**7, 0.25*10**7))
ax.plot(tpu_FC["# MACs"], tpu_FC["Inference time"], marker=".", label = "Edge TPU")
ax.plot(cpu_1FC["# MACs"], cpu_1FC["Inference time"], marker=".", label = "Intel Core i9-9900K 1 thread")
ax.plot(cpu_2FC["# MACs"], cpu_2FC["Inference time"], marker=".", label = "Intel Core i9-9900K 2 threads")
ax.plot(cpu_4FC["# MACs"], cpu_4FC["Inference time"], marker=".", label = "Intel Core i9-9900K 4 threads")
ax.plot(cpu_8FC["# MACs"], cpu_8FC["Inference time"], marker=".", label = "Intel Core i9-9900K 8 threads")

ax.set_xlabel("# MACs")
ax.set_ylabel("Tiempo medio de inferencia (ms)")
ax.legend()
plt.tight_layout(pad=0.3)
plt.grid(linestyle='--')
plt.savefig('TFMplots/FC_TPU_CPU_MAC.pdf')
plt.show()

cpu_8CONV = pd.read_csv("CONV_MACs/results/cpu-minF32-maxF702-stepF10-threads8.csv")
cpu_4CONV = pd.read_csv("CONV_MACs/results/cpu-minF32-maxF702-stepF10-threads4.csv")
cpu_2CONV = pd.read_csv("CONV_MACs/results/cpu-minF32-maxF702-stepF10-threads2.csv")
cpu_1CONV = pd.read_csv("CONV_MACs/results/cpu-minF32-maxF702-stepF10-threads1.csv")
tpu_CONV = pd.read_csv("CONV_MACs/results/minF32-maxF702-stepF10.csv")
fig,ax = plt.subplots()
plt.title("Tiempo en Edge TPU y CPU. (Capas de convolución)", pad=15, x=0.43)
ax.plot(tpu_CONV["# MACs"], tpu_CONV["Inference time"], marker=".", label = "Edge TPU")
ax.plot(cpu_1CONV["# MACs"], cpu_1CONV["Inference time"], marker=".", label = "Intel Core i9-9900K 1 thread")
ax.plot(cpu_2CONV["# MACs"], cpu_2CONV["Inference time"], marker=".", label = "Intel Core i9-9900K 2 threads")
ax.plot(cpu_4CONV["# MACs"], cpu_4CONV["Inference time"], marker=".", label = "Intel Core i9-9900K 4 threads")
ax.plot(cpu_8CONV["# MACs"], cpu_8CONV["Inference time"], marker=".", label = "Intel Core i9-9900K 8 threads")
ax.set_xlabel("# MACs")
ax.set_ylabel("Tiempo medio de inferencia (ms)")
ax.legend()
ax.set_yticks(np.arange(0, 4000, 500))
plt.tight_layout(pad=0.3)
plt.grid(linestyle='--')
plt.savefig('TFMplots/CONV_TPU_CPU_MAC.pdf')
plt.show()

print("Speedup 2 threads")
print(cpu_1CONV["Inference time"] / cpu_2CONV["Inference time"])
print("\nSpeedup 4 threads")
print(cpu_1CONV["Inference time"] / cpu_4CONV["Inference time"])
print("\nSpeedup 8 threads")
print(cpu_1CONV["Inference time"] / cpu_8CONV["Inference time"])
```
